* merge function, please read the document on https://pandas.pydata.org/pandas-docs/version/0.23.4/merging.html

```python
df1 = pd.DataFrame({'Country_Code': ['ARB', 'CSS', 'CEB', 'EAR'],
 'B': ['B0', 'B1', 'B2', 'B3'],
 'C': ['C0', 'C1', 'C2', 'C3'],
 'D': ['D0', 'D1', 'D2', 'D3']},
 index=[0, 1, 2, 3])
```

```python
df1.head()
```

```python
df2 = df.merge(df1,left_on='Country Code',right_on='Country_Code')
```
